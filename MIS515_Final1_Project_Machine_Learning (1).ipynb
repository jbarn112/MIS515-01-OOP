```python
# Load the dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

file_path = '/content/user_behavior_dataset.csv'
df = pd.read_csv(file_path)

# Display first few rows to understand the dataset structure
print(df.head())

df.describe()

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#DATA CLEANING PROCESS
# Show a summary of missing values in the dataset
missing_values_summary = df.isnull().sum()
print(missing_values_summary)

# Display total number of missing values in the dataset
total_missing_values = df.isnull().sum().sum()
print(f"Total missing values in the dataset: {total_missing_values}")

# Display rows with missing values
rows_with_missing_values = df[df.isnull().any(axis=1)]
print(rows_with_missing_values)


# Handle missing values
# Fill missing numerical values with mean
df.fillna(df.mean(numeric_only=True), inplace=True)

# Fill missing categorical values with mode
for col in df.select_dtypes(include=['object']).columns:
    df[col].fillna(df[col].mode()[0], inplace=True)

# Removing duplicates
df.drop_duplicates(inplace=True)

# Convert 'Screen On Time (hours/day)' to minutes per day
df['Screen On Time (minutes/day)'] = df['Screen On Time (hours/day)'] * 60

# HANDLING POOSIBLE OUTLIERS
# Identify Outliers Using Boxplots
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns

plt.figure(figsize=(15, 10))
for i, col in enumerate(numeric_columns, 1):
    plt.subplot(3, 3, i)
    sns.boxplot(y=col, data=df)
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()

#Interquartile Range (IQR) to detect outliers
for col in numeric_columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

 # Filter out the outliers
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

print("Dataset shape after removing outliers:", df.shape)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Data  Visualization

# Plot 1: Distribution of App Usage Time
plt.figure(figsize=(10, 6))
sns.histplot(df['App Usage Time (min/day)'], bins=20, kde=True, color='blue')
plt.title('Plot 1: Distribution of App Usage Time (min/day)')
plt.xlabel('App Usage Time (min/day)')
plt.ylabel('Frequency')
plt.show()


# Plot 2: Correlation Heatmap
plt.figure(figsize=(12, 8))
numeric_df = df.select_dtypes(include=[np.number])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Plot 2: Correlation Heatmap')
plt.show()

# Plot 3: Pairplot of Selected Features to identify relationships
selected_features = ['App Usage Time (min/day)', 'Age', 'Screen On Time (minutes/day)', 'Battery Drain (mAh/day)'] #add more variables
sns.pairplot(df[selected_features])
plt.show()

# Save the cleaned dataset
df.to_csv('/content/cleaned_user_behavior_dataset.csv', index=False)

# Display dataset info after cleaning to confirm changes
df.info()


```

```python
# Next step: Prepare data for Linear Regression
# Split the dataset into training and testing sets

#replacing gender with 1 and 0 to work with regression
df['Gender'] = df['Gender'].map({'Male': 1, 'Female': 0})

X = df[['Screen On Time (minutes/day)', 'Battery Drain (mAh/day)', 'Number of Apps Installed', 'Data Usage (MB/day)', 'Age']]
Y = df['App Usage Time (min/day)']

#train,Test split
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.2,random_state = 10)


#Training Model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,Y_train)

#intercept and Coefficients

print(model.intercept_)

coeff_df = pd.DataFrame(model.coef_,X.columns,columns=['Coefficient'])
coeff_df
```

```python
from matplotlib import pyplot as plt
coeff_df['Coefficient'].plot(kind='line', figsize=(8, 4), title='Coefficient')

# Get the current Axes instance
ax = plt.gca()

# Rotate x-axis labels for better readability
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')

# Remove top and right spines for cleaner look
ax.spines[['top', 'right']].set_visible(False)

# Show the plot
plt.show()
```

```python
#prediciton model set up
predictions = model.predict(X_test)
plt.scatter(Y_test,predictions)

#evalulation metrics

from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(Y_test, predictions))
print('MSE:', metrics.mean_squared_error(Y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(Y_test, predictions)))
```

With an MAE value of 33 we are able to tell the predictions of the model are off about 33 mins when it comes to predicting app usage time.

With a high MSE value of 1796.8 which indicates that the model is experiencing large errors in its predictions. This could be from penalizing outliers.

when taking the root for RMSE the errors look more in line with the data showing we are off in our predicitons by 42 mins.
